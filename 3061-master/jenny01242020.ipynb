```python
from scipy.integrate import odeint
import numpy as N

def f(y,t):
    return 9.81-(12.5/68.)*y

def f2(y,t):
    return 9.81-(12.5/68.)*y**2

def f3(y,t):
    return 9.81-(12.5/68.)*y**2*1*t


y0 = 0
a = 0
b = 12

t = N.arange(a,b,0.01)
y = odeint(f,y0,t)
y2= odeint (f2,y0,t)
y3= odeint (f3,y0,t)
import pylab
pylab.plot(t,y,t,y2,t,y3)
pylab.xlabel('t');pylab.ylabel('y(t)')
pylab.show()
```
